Problem Statement

Explain the difference between Absolute and Relative positioning.
-  # Absolute position- An element using position absolute will be positioned relative to the nearest ancestor this ancestor should be a positioned ancestor i.e.having a position value other than state.
position absolute removes the element from normal document flow.
   # Relative position- in position relative the element is by default placed according to the normal flow of the document.
   In a nutshell,relative sets the position of the element with respect to it's original position.

Task 5:


Problem Statement

Explain the z-index, with a code example.
- The z-index CSS property arranges the z-position of a positioned element, as well as its descendants.
The z-index property takes an integer value, which represents the element's position on the z-axis. Elements
with higher z-index values are placed on top of elements with lower z-index values.
By default, elements have a z-index value of 0, which means their stacking order is determined by their position
in the HTML document and their parent's stacking order.


```python
<! DOCTYPE html>
<html lang="en">
<head>
<title> About us</title>
<style>
.container {
    position:relative;
}
.box {
    position:absolute;
    width:100px;
    height:100px;
    font-size:20px;
    color:white;
    padding:5px;
}
.box--1 {
   background-color:red;
}
.box--2 {
    left:50px;
    top:50px;
    background-color:blue;
}
.box--3 {
    left:100px;
    top:100px;
    background-color:green;
}
</style>
</head>
<body>
<div class="container">
<div class="box box--1">z-index 1</div>
<div class="box box--2">z-index 2</div>
<div class="box box--3">z-index 3</div>
</div>
</body>
</html

```
